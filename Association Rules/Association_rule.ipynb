# Association Rule / Market Basket Analysis

```python
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')
from mlxtend.frequent_patterns import apriori, association_rules
from mlxtend.preprocessing import TransactionEncoder
```

```python
df=pd.read_excel("Online retail.xlsx",header = None)
df
```

```python
def textsplit(txt):
    return txt.split(',')
```

```python
df=df[0].apply(textsplit)
```

```python
df
```

```python
te = TransactionEncoder()
encoded_df = te.fit_transform(df)
encoded_df
```

```python
te.columns_
```

```python
data = pd.DataFrame(encoded_df,columns=te.columns_)
data
```

```python
data.replace([True,False],[1,0],inplace=True)
data
```

```python
scores = apriori(data,min_support=0.01,use_colnames=True)
scores
```

```python
rules= association_rules(scores,num_itemsets=257,min_threshold=0.2)
rules
```

```python
rules[rules['lift']>1]
```
